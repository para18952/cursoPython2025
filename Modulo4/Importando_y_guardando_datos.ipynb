Consideraciones:
1. A pesar de la estructura tabular de Excel, que invita a organizar los datos en filas y columnas,
   las hojas de c√°lculo no son conjuntos de datos y no exigen que la gente almacene los datos de esa manera.

2. Suele haber informaci√≥n adicional en filas o columnas antes o despu√©s de los datos que hay que importar.

3. Los tipos de datos no siempre est√°n tan claros como para la persona que cre√≥ la hoja de c√°lculo.

4. Excel no insiste en que todos los datos de una columna sean del mismo tipo, o que los t√≠tulos de las columnas sean apropiados para su
   uso con un lenguaje de programaci√≥n como Python.

```python
def gettots(x):
    out = {}
    out['min'] = x.min()
    out['qr1'] = x.quantile(0.25)
    out['med'] = x.median()
    out['qr3'] = x.quantile(0.75)
    out['max'] = x.max()
    out['count'] = x.count()
    return pd.Series(out)
```
