### Operador Ternario

### Ejercicio 1:
- Transforma el siguiente condicional usando el operador ternario:

```python
valor = random.randint(1000, 2000)

if valor % 2 == 0:
    
    x = "par"
    
else:
    
    x = "impar"
    
print(x)
```

### Ejercicio 2:
- Transforma el siguiente condicional usando el operador ternario:

```python
a = random.random()
b = random.random()

if a > b:
    x = a
else:
    x = b
    
print("a:", a)
print("b:", b)
print("x:", x)
```

### Ejercicio 3:
- Transforma el siguiente condicional usando el operador ternario:

```python
string = input("Ingresa una frase con palabras separadas por espacios: ")

if len(string.split()) > 5:
    x = len(string.split())
    
else:
    x = 0
    
print(x)
```

### Ejercicio 4:
- Transforma el siguiente condicional usando el operador ternario:

```python
numero = 5

if type(numero) == int:
    x = numero**2
    
else:
    x = numero
    
print(x)
```
