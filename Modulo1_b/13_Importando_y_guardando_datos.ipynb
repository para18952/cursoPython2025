```python
# Para contar valores únicos y duplicados en cada columna
total_total = 0
for col in landtemps.columns:                                                                # <- Recorremos todas las columnas
  # <- Guardammos en una variable el número de valores unicos por cada columna
  unicos = landtemps[col].nunique()
  total_total += unicos
  # <- Guardammos en una variable el número total de valores
  total_valores = landtemps[col].count()
  # <- Guardammos en una variable el número de valores duplicados
  duplicados = total_valores - unicos
  print(f"Columna '{col}': {duplicados} valores duplicados ({unicos} valores únicos)")     # <- Imprimimos el resultado
print(f"\nTotal de valores únicos: {total_total}")
```
