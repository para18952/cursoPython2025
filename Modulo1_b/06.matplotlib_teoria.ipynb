## Matplotlib

**`matplotlib`** es una biblioteca para la generación de gráficos a partir de datos contenidos en **`listas`** o **`arrays`**.

```python
import matplotlib.pyplot as plt
```

Para generar las gráficas más comunes tenemos los siguientes métodos:

|Función            | Plot                         |
|-------------------|------------------------------|
|**`plt.plot()`**   | Grafica **X vs Y** (lineal). |
|**`plt.scatter()`**| Graficas de dispersión.      |
|**`plt.bar()`**    | Grafico de barras vertical.  |
|**`plt.barh()`**   | Grafico de barras horizontal.|


Graficos estadísticos:

|Función               |Plot               |
|----------------------|-------------------|
|**`plt.hist()`**      | Histogramas.      |
|**`plt.boxplot()`**   | Grafico de cajas. |
|**`plt.violinplot()`**| Grafico de violin.|
|**`plt.pie()`**       | Diagrama circular.|

````python
import numpy as np
import matplotlib.pyplot as plt
````

### plt.plot()

**`plt.plot`** hace gráficas de **X vs Y**, por lo que necesitamos una lista de valores para **X** y una para **Y**.

````python
# En este ejemplo vamos a generar 2 arreglos de datos: 1 con np.linspace() y otra aleatoria

# Generar un array de datos igualmente distribuidos de entrada
X = np.linspace(0, 10, 100)

# Generar un array de datos de salida
Y = np.exp(-X) * np.cos(2*np.pi*X)
````

````python
plt.plot(X, Y)
plt.show()
````

### Personalizar plots

La libreria **`matplotlib`** tiene muchas formas de personalizar los plots, tanto usando los parámetros de la funciones o con otras funciones:

|Función                                  |Descripción                                          |
|-----------------------------------------|-----------------------------------------------------|
|**`plt.figure(figsize = (alto, ancho))`**| Modifica el tamaño del plot.                        |
|**`plt.title()`**                        | Agrega un titulo al plot.                           |
|**`plt.xlabel()`**                       | Agrega una etiqueta al eje X.                       |
|**`plt.ylabel()`**                       | Agrega una etiqueta al eje Y.                       |
|**`plt.legend()`**                       | Si se pasaron etiquetas al plot, muestra la leyenda.|
|**`plt.xtick()`**                        | Modifica la vista de las marcas en el eje X.        |
|**`plt.ytick()`**                        | Modifica la vista de las marcas en el eje Y.        |
|**`plt.grid()`**                         | Agrega una malla o **"grid"** al plot.              |

### Personalizar plt.plot()

La función **`plt.plot()`** tiene varios parámetros que podemos utilizar para agregar colores y estilos al plot.

````python
plt.plot(
        X,Y,
        color = "Red",
        label = "Exponencial",
        linewidth = 0.5,
        marker= "o",
        markersize = 0.8
         )

plt.xlabel("X", fontsize = 15, labelpad = 20)
plt.ylabel("Y", fontsize = 15, labelpad = 20)
plt.title("Gráfico de exponencial", fontsize = 18, fontweight = "bold", fontstyle="italic", pad = 20)
plt.legend()
plt.show()
````

````python
# Vamos a graficar dos funciones.
# plt.figure(figsize=(8, 4))
# plt.tight_layout()
# fontweight=100–900
# fontstyle: italic,oblique,normal
# Esto permite aproximar las soluciones como  log(1/x) = sen(x^3)

X = np.linspace(1, 100000, 200)

# Plot de log(1/x)
Y1 = np.log(1/X)
plt.plot(X, Y1, label="log(1/x)")

# Plot de sen(x^3)
Y2 = np.sin(X**3)
plt.plot(X, Y2, label="sen(x^3)")

plt.title("Gráfico de las funciones $log(\\frac{1}{x})$ y $x^3$", fontsize = 18, fontweight = "bold", fontstyle="italic", pad = 20)
plt.legend( bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
````

### plt.scatter()

**`plt.scatter()`** hace gráficas de dispersion de **X** vs **Y**, por lo que necesitamos una lista de valores para **X** y una para **Y**.

````python
# Vamos a generar 2 arreglos de datos
X = np.linspace(0, 10, 100)
ruido = np.random.random(X.size)
Y = np.sin(X)+ ruido/12
Y2 = np.sin(X)

#plt.scatter(X, Y)
plt.scatter (
              X,
              Y,
              label="$sen(x^3)$",
              marker="<",
              s=40,
              edgecolors="blue",
              facecolors="green",
              linewidth=1,
              alpha = 0.5
            )
plt.plot(X, Y2, color = "red", linewidth = 1)
plt.legend()
plt.show()
````

### plt.bar() y plt.barh()
**`plt.bar()`** y **`plt.barh()`** hacen gráficas de barras de **X** con una lista de etiquetas, por lo que necesitamos una lista de valores para **X** y una lista de etiquetas.

````python
# Voy a generar una lista aleatoria

# plt.bar()
# Vamos a simular un conteo de casos donde se cuenta el número de personas que ganan mas de 1000 euros

np.random.seed(5)
X = np.random.randint(0, 3000, 6)

etiquetas = ["Cordoba", "Valladolid", "La Coruña", "Móstoles", "Alicante","Almería"]

plt.bar(etiquetas, X,
        width=0.5,
        align= "center",
        edgecolor = "yellow",
        linewidth = 2,
        #color = "blue",
        alpha = 0.5,
        label = "Personas que ganan más de 1000 euros"
        ) # que la quiero horizontal, se le tiene (h)
plt.legend()
plt.show()
````

### Subplots

**`subplot`** es una función para hacer graficas en una **"malla"** o **"matriz"**.

````python
fig, ax = plt.subplots(figsize=(6, 4))      # fig = hoja, ax = recuadro de dibujo

X = np.linspace(0, 10, 2000)                # Generar un array de datos igualmente distribuidos de entrada
Y = np.exp(-X) * np.cos(2*np.pi*X)          # Generar un array de datos de salida

ax.plot(X, Y)                               # Dibujar en ese eje
ax.set_title("Gráfico en el eje")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")

fig.suptitle("Título de toda la figura")    # Título general encima de todo
fig.tight_layout()                          # Ajusta márgenes automáticamente

plt.savefig("figura_1.png", dpi=300)
plt.show()
````

````python
# Creamos la figura "fig" donde se le van agregar 3 subplots utilizando la variable "ax"
fig, ax = plt.subplots(figsize=(15, 6), nrows=1, ncols=3)

################# PRIMERA ##########################
XX = np.linspace(0, 10, 2000)                # Generar un array de datos igualmente distribuidos de entrada
YY = np.exp(-XX) * np.cos(2*np.pi*XX)          # Generar un array de datos de salida

ax[0].plot(XX, YY)                               # Dibujar en ese eje
ax[0].set_title("Gráfico en el eje")
ax[0].set_xlabel("Eje X")
ax[0].set_ylabel("Eje Y")


################# SEGUNDA ##########################
np.random.seed(5)
X = np.random.randint(0, 3000, 6)
etiquetas = ["Cordoba", "Valladolid", "La Coruña", "Móstoles", "Alicante","Almería"]
ax[2].barh(etiquetas, X,
        height=0.5,
        align= "center",
        edgecolor = "yellow",
        linewidth = 2,
        #color = "blue",
        alpha = 0.5,
        label = "Personas contaminadas"
        )
ax[2].set_xlabel("Personas")
ax[2].set_ylabel("Ciudades")
ax[2].set_title("Personas contaminadas")
ax[2].legend()


################# TERCERA ##########################
X2 = np.linspace(0, 10, 100)
ruido = np.random.random(X2.size)
Y = np.sin(X2)+ ruido/12
Y2 = np.sin(X2)

#plt.scatter(X, Y)
ax[1].scatter (
              X2,
              Y,
              label="$sen(x^3)$",
              marker="<",
              s=40,
              edgecolors="blue",
              facecolors="green",
              linewidth=1,
              alpha = 0.5
            )
ax[1].plot(X2, Y2, color = "red", linewidth = 1)
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")
ax[1].set_title("Gráfico de dispersión")
ax[1].legend()


####################################################

fig.suptitle("Título de toda la figura")    # Título general encima de todo
fig.tight_layout()                          # Ajusta márgenes automáticamente

plt.savefig("figura_1.png", dpi=300)
plt.show()
````

## 3D

````python
# Grafica de la función de Schwefel
# Crear una malla 2D para X y Y
x = np.linspace(-100, 100, 100)
y = np.linspace(-100, 100, 100)
X, Y = np.meshgrid(x, y)

# calcular la función Z
Z = 418.9829 * 2 - (X * np.sin(np.sqrt(np.abs(X))) + Y * np.sin(np.sqrt(np.abs(Y))))

flg, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10, 10))
#ax.plot_surface(X, Y, Z, cmap="viridis")
ax.plot_surface(X, Y, Z, alpha=0.9)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
````

#### Boxplot

**`plt.boxplot()`** para los boxplots necesitamos una lista de datos para cada categoria. Muestra un resumen de una gran cantidad de datos en cinco medidas descriptivas, además de intuir su morfología y simetría. Este tipo de gráficos nos permite identificar valores atípicos (outliers) y comparar distribuciones.

Los datos se dividen en cuatro partes iguales, separadas por puntos llamados cuartiles (quartiles).

- Rango (sin datos atípicos)
- Datos atípicos (outliers)
- Rango Intercuartil (también conocido como RIC)
- Cuartiles (Q1, Q2 y Q3)
- Mediana (Q2)
- Mínimo y Máximo
